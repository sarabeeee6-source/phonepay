```python
# %%
import streamlit as st
import pandas as pd
import sqlalchemy
import plotly.express as px

# -----------------------------
# Database connection settings
# -----------------------------
# Example for MySQL; change URL for PostgreSQL, SQL Server, etc.
DB_URL = "mysql+pymysql://root:root@localhost:3306/phonepay"

# Create SQLAlchemy engine
try:
    engine = sqlalchemy.create_engine(DB_URL)
except Exception as e:
    st.error(f"Database connection failed: {e}")
    st.stop()

# -----------------------------
# Fetch data from SQL
# -----------------------------
import pymysql

# Connect to MySQL
conn = pymysql.connect(
    host="localhost",
    user="root",
    password="root",
    database="phonepay"
)

# --- Create table ---
cursor = conn.cursor()
cursor.execute( """
SELECT 
    State,
    Year,
    SUM(Transaction_Count) AS total_transaction_count,
    SUM(Transaction_Amount) AS total_transaction_amount
FROM agg_transaction
GROUP BY State, Year
ORDER BY Year;
""")

results = cursor.fetchall()
cursor.close()
conn.close()

# -----------------------------
# Build DataFrame
# -----------------------------
df = pd.DataFrame(
    results,
    columns=['State', 'Year', 'total_transaction_count', 'total_transaction_amount']
)

# -----------------------------
# Streamlit UI
# -----------------------------
st.title("Transactions Dashboard")

# Add "All States" option
state_options = ["All States"] + sorted(df["State"].unique().tolist())
selected_state = st.selectbox("Filter by State (optional):", state_options)

# -----------------------------
# Filter data
# -----------------------------
if selected_state == "All States":
    # Aggregate across all states
    agg_df = df.groupby("Year", as_index=False).agg({
        "total_transaction_count": "sum",
        "total_transaction_amount": "sum"
    })

    # Plot totals
    fig_count = px.line(
        agg_df,
        x="Year",
        y="total_transaction_count",
        markers=True,
        title="Year-wise Total Transaction Count (All States)"
    )
    st.plotly_chart(fig_count, width="stretch")

    fig_amount = px.line(
        agg_df,
        x="Year",
        y="total_transaction_amount",
        markers=True,
        title="Year-wise Total Transaction Amount (All States)"
    )
    st.plotly_chart(fig_amount, width="stretch")

else:
    # Show plots only for the selected state
    filtered_df = df[df["State"] == selected_state]

    fig_count = px.line(
        filtered_df,
        x="Year",
        y="total_transaction_count",
        markers=True,
        title=f"Year-wise Transaction Count for {selected_state}"
    )
    st.plotly_chart(fig_count, width="stretch")

    fig_amount = px.line(
        filtered_df,
        x="Year",
        y="total_transaction_amount",
        markers=True,
        title=f"Year-wise Transaction Amount for {selected_state}"
    )
    st.plotly_chart(fig_amount, width="stretch")
```
